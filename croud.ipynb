```python
import random
import pandas as pd
import numpy as np
```

```python
data = pd.read_html('https://bmstu.ru/bachelor/previous-points?ysclid=lgpdodbo5c867324')
```

```python
df = data[0]
```

```python
df
```

```python
df.drop(columns=['Код НП(С)', 'Факультет', 'Кафедра'], inplace=True)
```

```python
df.rename(columns = {'Наименование направления подготовки (специальности) – НП(С)':'direction', 'Квалификация':'qualification'}, inplace=True)
```

```python
df
```

```python
direction_df =  df.xs('direction', axis=1, level=1)
qualification_df = df.xs('qualification', axis=1, level=1)
df_2022 = df[('2022', 'ОК')]
df_2021 = df[('2021', 'ОК')]
df_2020 = df[('2020', 'ОК')]
df_2019 = df[('2019', 'ОК')]
df_2018 = df[('2018', 'ОК')]
df_2017 = df[('2017', 'ОК')]

new_df = pd.DataFrame(pd.concat([direction_df, qualification_df, df_2022, df_2021, df_2020, df_2019, df_2018, df_2017], axis=1))
new_df.columns = ['direction', 'qualification', '2022', '2021', '2020', '2019', '2018', '2017']
new_df
```

```python
new_df.drop(new_df[new_df['qualification'] == 'Специалист'].index, inplace=True)
new_df.reset_index(drop=True, inplace=True)
new_df
```

```python
directions = ['Автоматизация технологических процессов и производств', 'Бизнес-информатика', 'Информатика и вычислительная техника', 
              'Информационные системы и технологии', 'Математика и компьютерные науки', 'Прикладная информатика', 'Прикладная математика',
              'Прикладная математика и информатика', 'Ракетные комплексы и космонавтика']
new_df['direction'].unique()
```
